```python
#  Convertendo colunas
df["Data"] = pd.to_datetime(df["Data"])
df["Turno Inicio"] = pd.to_datetime(df["Turno Inicio"], format="%H:%M:%S")
df["Turno Fim"] = pd.to_datetime(df["Turno Fim"], format="%H:%M:%S")
df["Hora Inicio"] = pd.to_datetime(df["Hora Inicio"], format="%H:%M:%S")
df["Hora Fim"] = pd.to_datetime(df["Hora Fim"], format="%H:%M:%S")
df["Máquina"] = df["Máquina"].astype("string")
df["Motivo Parada"] = df["Motivo Parada"].astype("string")
df["Ferramenta"] = df["Ferramenta"].astype("string")
```

```python
df.dtypes
```

```python
#  Criando coluna calculada de Total de Horas paradas
df["Parada"] = df["Hora Fim"] - df["Hora Inicio"]
```
